```python
import sys  
sys.path.insert(0, '/root/pgp')

%load_ext autoreload
%autoreload 2
from train_eval.initialization import initialize_dataset, get_specific_args

import yaml

CONFIG = "/root/pgp/configs/pgp_gatx2_lvm_traversal.yml"
DATA_ROOT = "/nuscenes_CLEAN"
DATA_DIR = "/nuscenes_PGP_PREPROCESS/visuals"


# Load config
with open(CONFIG, 'r') as yaml_file:
    cfg = yaml.safe_load(yaml_file)

# Initialize dataset
ds_type = cfg['dataset'] + '_' + cfg['agent_setting'] + '_' + cfg['input_representation']
spec_args = get_specific_args(cfg['dataset'], DATA_ROOT, cfg['version'] if 'version' in cfg.keys() else None)
test_set = initialize_dataset(ds_type, ['load_data', DATA_DIR, cfg['test_set_args']] + spec_args)
```

```python
import cv2
from matplotlib import pyplot as plt

def paint_view(img, view):
    min_x = int(view['roi_min_x'])
    min_y = int(view['roi_min_y'])
    max_x = int(view['roi_max_x'])
    max_y = int(view['roi_max_y'])
    cv2.rectangle(img, (min_x, min_y), (max_x, max_y), (255, 0, 0), 2)

def show_imgs(imgs, n_col=5, dpi=100):
    # Determine the number of images and the number of rows and columns in the grid
    n_images = len(imgs)
    n_cols = n_col
    n_rows = (n_images + n_cols - 1) // n_cols

    # Create a figure and a grid of subplots with the desired number of rows and columns
    fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(50, 50), dpi=dpi)

    # Iterate over the subplots and the images
    for ax, img in zip(axs.flat, imgs):
        # Display the image in the current subplot
        ax.imshow(img)

    # Remove the axis labels
    for ax in axs.flat:
        ax.axis('off')

    plt.show()

def show_dp_visuals(dp):
    visuals = dp['inputs']['target_agent_visuals']
    imgs = []
    for view in visuals:
        img = cv2.imread(f"{DATA_ROOT}/{view['cam_file']}")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        paint_view(img, view)
        imgs.append(img)
    show_imgs(imgs, 5, 100)
    
```
